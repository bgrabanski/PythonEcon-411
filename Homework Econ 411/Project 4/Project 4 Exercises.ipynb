```python
gdpdata = pd.read_csv("GDP.csv")
gdpdata
```

```python
gdpgrowth = gdpdata[0:5]
gdpgrowth
```

```python
gdpcompare = gdpdata.loc[1:303]
```

```python
gdpdata["$ Change"] = gdpdata["GDP"].diff()
gdpdata["% Change"] = gdpdata["GDP"].pct_change()
gdpdata
```

```python

gdpdata_yearly = gdpdata.loc[::4]
gdpdata_yearly["Year"] = gdpdata_yearly["DATE"].str[:4]
gdpdata_yearly1 = gdpdata_yearly.drop(["% Change","DATE","$ Change"],axis = 1)
gdpdata_yearly1
```

```python
gdpdata_yearly1["% Change"] = gdpdata_yearly1["GDP"].pct_change()
gdpdata_yearly1["$ Change"] = gdpdata_yearly1["GDP"].diff()
gdpdata_yearly1 = gdpdata_yearly1.fillna(0).set_index("Year")
gdpdata_yearly1
```

```python
for i in range(0,304):
    if i != 0:
        print(gdpdata["DATE"][i], gdpdata["GDP"][i], gdpdata["% Change"][i], gdpdata["$ Change"][i])
    else:
        print( gdpdata["DATE"][i], gdpdata["GDP"][i],"Unvavailable","        Unavailable")
```

```python
import matplotlib.pyplot as plt
```

```python
gdpdata_yearly1.plot.line(legend = "best")
```

```python
plt.rcParams.update({"font.size" : 16})
fig, ax = plt.subplots(figsize = (25, 16))
plot_df = gdpdata_yearly1.dropna()
plot_df.plot.bar(y = ["GDP", "% Change"],
                secondary_y = "% Change",
                fill = True,
                alpha= .7,
                width = .95,
                 legend = True,
                ax = ax,
                color = "C5",
                linewidth = 12)
ax.set_yticklabels(["$" + str(y) for y in ax.get_yticks()])
x_ticklabels = [ str(x) for x in plot_df.index]
ax.set_xticklabels(x_ticklabels)

ax.set_title("GDP and GDP Growth Over Time")
```
