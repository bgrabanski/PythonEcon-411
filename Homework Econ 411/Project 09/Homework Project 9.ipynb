```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python

hw_data = pd.read_excel("EFWAndRGDP.xls")
```

```python


hw_data.reset_index(inplace = True)
hw_data["Year"] = hw_data["Year"].astype(str).astype("datetime64[ns]").sort_index()
hw_data = hw_data.set_index(["ISO_Code_3", "Year"]).sort_index()
del hw_data["index"]
del hw_data["ISO_Code_2"]
del hw_data["World Bank Region"]
del hw_data["World Bank Current Income Classification, 1990-present (L=Low income, LM=Lower middle income, UM=Upper middle income, H=High income)"]
hw_data
```

```python
reg_df = hw_data[hw_data["Countries"].isin(oecd_countries)]
reg_df
```

```python
## Regressions with Logged Differences
years_diff = 5
r_df = reg_df
# take the log of real gdp then group by difference
r_df["RGDP Per Capita"] = np.log(reg_df["RGDP Per Capita"]).groupby(
    "ISO_Code_3").diff(years_diff)

# replace infinite values with null values since it will cause issues
r_df = r_df.replace([np.inf, -np.inf], np.nan)
r_df
```

```python
clean_df = r_df.dropna(axis = 0, how = "any")
#clean_df = clean_df[y_var + x_vars].groupby("ISO_Code_3").diff(years_diff)


# create variables
y_var = ["RGDP Per Capita"]
x_vars = ['Size of Government',
  'Legal System and Property Rights',
  'Sound Money',
  'Freedom to Trade Internationally',
  'Regulation']
y = r_df[y_var]
X = r_df[x_vars]
clean_df
```

```python
y = clean_df[y_var]
X = clean_df[x_vars]
X["Constant"] = 1
results = sm.OLS(y, X).fit()
clean_df["Predictor"] = results.predict()
results.summary()
```
